```python
import h5py
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
hy_path = '/fast_scratch_2/aferreira/t2k/ml/data/oct20_combine_flatE/' 
npy_path = '/fast_scratch_2/aferreira/t2k/ml/data/oct20_combine_flatE/nov29_normed_regression_shared/20092023-101855/'
idx = np.array(sorted(np.load(npy_path + "/indices.npy")))

hy = h5py.File(hy_path+'combine_combine.hy', "r")
print(hy.keys()) # find hit time thing

hit_time = np.array(hy['hit_time']) # why is this still not a 2d shape?
event_hits_index = np.array(hy['event_hits_index'])

#[idx].squeeze() 
print(hit_time.shape) # should be 2d? 
print(hit_time[0])
#plt.show()  
```

```python
events_to_plot = 5

plt.figure(figsize=(10,7))
plt.title(f'hit time distrubition for first {events_to_plot} events in HDF5 file \n plotted seperately with automatic binning', fontsize=15)


for i in range(events_to_plot): # first 10 from overall dataset
    hit_times_i = hit_time[event_hits_index[i]:event_hits_index[i+1]]
    plt.hist(hit_times_i, bins='auto', alpha=0.3, label=f'event {i}') # bins=10 didnt work

plt.xlabel('hit_time')
plt.ylabel('count')

plt.legend()
plt.show()
```

```python
events_to_plot = 10000
plt.title(f'hit time distrubition for first {events_to_plot} events in HDF5 file \nplotted together with automatic binning')
hit_times_i = hit_time[event_hits_index[0]:event_hits_index[events_to_plot+1]]

plt.hist(hit_times_i, bins='auto', alpha=0.5) # bins=10 didnt work

plt.xlabel('hit_time')
plt.ylabel('count')

plt.legend()
plt.show()
```
